```python
data.drop(columns=['Review'], axis = 1)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
train, test = train_test_split(data, test_size=0.3, random_state=2021)
train.shape, test.shape
```

```python
x_train= train.drop(['Date','MaintenanceTask','Weekday','Holiday','Temperature','Humidity','Windy','Crowdedness','ID','Play','PlayTimeHour','Review'],1)
y_train= train['MaintenanceTask']
x_test = test.drop(['Date','MaintenanceTask','Weekday','Holiday','Temperature','Humidity','Windy','Crowdedness','ID','Play','PlayTimeHour','Review'],1)
y_test = test['MaintenanceTask']
```

```python
x_train
```

```python
y_train
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le_Month=LabelEncoder()
le_Season=LabelEncoder()
le_Outlook=LabelEncoder()
le_EmailCampaign=LabelEncoder()
```

```python
x_train['Month'] = le_Season.fit_transform(x_train['Month'])
x_train['Season'] = le_Season.fit_transform(x_train['Season'])
x_train['Outlook'] = le_Season.fit_transform(x_train['Outlook'])
x_train['EmailCampaign'] = le_Season.fit_transform(x_train['EmailCampaign'])
```

```python
x_train.head()
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
model.fit(x_train, y_train)
model.score(x_train,y_train)
```
