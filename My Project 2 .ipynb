```python
MODELLING
```

```python
from sklearn.linear_model import LinearRegression
```

```python
from sklearn.model_selection import train_test_split
```

```python
train, test = train_test_split(data, test_size=0.3, random_state=2021)
```

```python
train.shape, test.shape
```

```python
x_train = train.drop(['Temperature','Date','Month','Weekday','Crowdedness','Holiday','MaintenanceTask','EmailCampaign','ID','Play','PlayTimeHour','Review'],1)
y_train = train['Temperature']
```

```python
x_test = test.drop(['Temperature','Date','Month','Weekday','Holiday','Crowdedness','MaintenanceTask','EmailCampaign','ID','Play','PlayTimeHour','Review'],1)
y_test = test['Temperature']
```

```python
lr = LinearRegression()
```

```python
lr.fit(x_train, y_train)
```

```python
c = lr.intercept_
```

```python
c
```

```python
lr.coef_
```

```python
lr.predict([[3,40.0,1,0]])
```

```python
x_test[2:3]
```

```python
y_predict_train = lr.predict(x_train)
```

```python
y_predict_train
```

```python
import matplotlib.pyplot as plt
plt.scatter(y_train,y_predict_train)
plt.xlabel('Actual lowest_salary')
plt.ylabel('Predicted lowest_salary')
plt.show()
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_train, y_predict_train)
```
